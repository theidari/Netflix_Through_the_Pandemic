```python
import pandas as pd

df = pd.read_csv(r'/content/abcdata.csv')
df
```

```python

import matplotlib.pyplot as plt
plt.scatter(df['Date'],df['Open'],color = 'g',s = 100)
plt.xticks(rotation = 25)
plt.xlabel('Date')
plt.ylabel('Open')
```

```python

fig = plt.figure()
ax1 = fig.add_subplot()
ax1.scatter(df['Date'], df['Open'], s=10, c='b', marker="s", label='Open')
plt.xlabel('Date')
plt.ylabel('Open')
plt.legend(loc='upper left')
plt.show()
```

```python
fig2 = plt.figure()
ax2 = fig2.add_subplot()
ax2.scatter(df['Date'],df['High'], s=10, c='r', marker="o", label='High')
plt.xlabel('Date')
plt.ylabel('High')
plt.legend(loc='upper left')
plt.show()
```

```python
fig3 = plt.figure()
ax3 = fig3.add_subplot()
ax3.scatter(df['Date'],df['Low'], s=10, c='y', marker="o", label='Low')
plt.xlabel('Date')
plt.ylabel('Low')
plt.legend(loc='upper left')
plt.show()
```

```python
fig4 = plt.figure()
ax4= fig4.add_subplot()
ax4.scatter(df['Date'],df['Close'], s=10, c='g', marker="o", label='Close')
plt.xlabel('Date')
plt.ylabel('Close')
plt.legend(loc='upper left')
plt.show()
```

```python
fig5 = plt.figure()
ax5 = fig5.add_subplot()
ax5.scatter(df['Date'],df['Adj Close'], s=10, c='#1f77b4', marker="o", label='Adj Close')
plt.xlabel('Date')
plt.ylabel('Adj Close')
plt.legend(loc='upper left')
plt.show()
```

```python
fig6 = plt.figure()
ax6 = fig6.add_subplot()
ax6.scatter(df['Date'],df['Volume'], s=10, c='#9467bd', marker="o", label='Volume')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend(loc='upper left')
plt.show()
```

```python
coor_p=df[['Open',	'High',	'Low'	, 'Close','Volume']].corr(method='pearson')
print("Correlation between all variables\n",coor_p)
```

```python
import seaborn as sns
sns.heatmap(coor_p)
```
